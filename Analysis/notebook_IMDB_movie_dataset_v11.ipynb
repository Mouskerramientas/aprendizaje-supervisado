```python
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns
```

```python
df = pd.read_csv("TMDB_movie_dataset_v11.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(14,6))
msno.matrix(df, labels=True, color=(0.2, 0.4, 0.6), fontsize=12)
plt.xticks(rotation=45, ha="right")  # diagonal como en tu imagen
plt.tight_layout()
plt.show()
```

```python
df.dropna(subset=['genres'], inplace=True)
```

```python
df.drop(['backdrop_path', 'homepage', 'imdb_id', 'overview', 'tagline', 'poster_path', 'production_companies', 'production_countries','spoken_languages', 'spoken_languages', 'keywords'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(14,6))
msno.matrix(df, labels=True, color=(0.2, 0.4, 0.6), fontsize=12)
plt.xticks(rotation=45, ha="right")  # diagonal como en tu imagen
plt.tight_layout()
plt.show()
```

```python
df.hist(figsize=(12, 8), bins=20, color='skyblue')
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns.drop('id')
```

```python
sns.boxplot(data=df[numeric_cols], showfliers=False)
```

```python
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col], showfliers=False)
    plt.title(f"Boxplot de {col}")
    plt.tight_layout()
    plt.show()
```
