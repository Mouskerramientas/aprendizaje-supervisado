```python
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns
```

```python
df = pd.read_csv("TMDB_movie_dataset_v11.csv")
```

```python
df.dropna(subset=['genres'], inplace=True)
```

```python
df.drop(['backdrop_path', 'homepage', 'imdb_id', 'overview', 'tagline', 'poster_path', 'production_companies', 'production_countries','spoken_languages', 'spoken_languages', 'keywords'], axis=1, inplace=True)
```

```python
df.info()
```

```python
plt.figure(figsize=(14,6))
msno.matrix(df, labels=True, color=(0.2, 0.4, 0.6), fontsize=12)
plt.xticks(rotation=45, ha="right")  # diagonal como en tu imagen
plt.tight_layout()
plt.show()
```

```python
df.dropna(subset=['release_date'], inplace=True)
```

```python
df.dropna(subset=['title'], inplace=True)
```

```python
df.dropna(subset=['original_title'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv('IMDB_cleaned.csv', index=False)
```
